```python
import torch
from torch import nn
from torch.nn import functional as F

from typing import Optional

from dataclasses import dataclass, replace
import numpy as np
import einops

from tqdm.notebook import trange

import time
import pandas as pd


import matplotlib.pyplot as plt
```

```python
save_dir = "/Users/jacklindsey/superposition_data"
```

```python
@dataclass
class Config:
  n_features: int
  n_neurons: int
  n_layers: int
  n_layers_ground_truth: int
  n_features_to_fit: int
  

  # We optimize n_instances models in a single training loop
  # to let us sweep over sparsity or importance curves 
  # efficiently.

  # We could potentially use torch.vmap instead.
  n_instances: int
 
class Model(nn.Module):
  def __init__(self, 
               config, 
               feature_probability: Optional[torch.Tensor] = None,
               importance: Optional[torch.Tensor] = None,               
               device='cuda'):
    super().__init__()
    self.config = config
    self.encoder = nn.Parameter(torch.empty((config.n_instances, config.n_features, config.n_neurons), device=device))
    self.decoder = nn.Parameter(torch.empty((config.n_instances, config.n_neurons, config.n_features_to_fit), device=device))

    self.W = []
    self.b = []
    for i in range(config.n_layers):
        self.W.append(nn.Parameter(torch.empty((config.n_instances, config.n_neurons, config.n_neurons), device=device)))
        self.b.append(nn.Parameter(torch.zeros((config.n_instances, config.n_neurons), device=device)))


    self.W_gt = []
    for i in range(config.n_layers_ground_truth):
        self.W_gt.append(nn.Parameter(torch.empty((config.n_instances, config.n_features, config.n_features), device=device)))
        self.W_gt[i].requires_grad = False
        
    self.out_gt = nn.Parameter(torch.empty((config.n_instances, config.n_features, config.n_features_to_fit), device=device))
    
    
    nn.init.xavier_normal_(self.encoder)
    nn.init.xavier_normal_(self.decoder)
    
    for i in range(config.n_layers):
        nn.init.xavier_normal_(self.W[i])
        
    for i in range(config.n_layers_ground_truth):
        nn.init.xavier_normal_(self.W_gt[i])
        
    nn.init.xavier_normal_(self.out_gt)
    self.b_final = nn.Parameter(torch.zeros((config.n_instances, config.n_features_to_fit), device=device))

    if feature_probability is None:
      feature_probability = torch.ones(())
    self.feature_probability = feature_probability.to(device)
    if importance is None:
      importance = torch.ones(())
    self.importance = importance.to(device)

  def forward(self, features):
    # features: [..., instance, n_features]
    # W: [instance, n_features, n_neurons]
    hidden = torch.einsum("...if,ifh->...ih", features, self.encoder)
    #hidden = F.relu(hidden)
    for i in range(self.config.n_layers):
        hidden = torch.einsum("...if,ifh->...ih", hidden, self.W[i]) + self.b[i]
        hidden = F.relu(hidden)
    out = torch.einsum("...if,ifh->...ih", hidden, self.decoder) + self.b_final
    out = F.relu(out) 
    return out


  def ground_truth(self, features):
    # features: [..., instance, n_features]
    # W: [instance, n_features, n_neurons]
    out = features
    for i in range(self.config.n_layers_ground_truth):
        out = torch.einsum("...if,ifh->...ih", out, self.W_gt[i])
        out = F.relu(out)
    #torch.einsum("...if,ifh->...ih", out, self.out_gt)
    out = out[:, :, :self.config.n_features_to_fit]
    return out

  def generate_batch(self, n_batch):
    feat = F.relu(torch.rand((n_batch, self.config.n_instances, self.config.n_features), device=self.encoder.device))
    #mask = np.zeros([n_batch, self.config.n_instances, self.config.n_features])
    #for b in range(n_batch):
    #    for i in range(self.config.n_instances):
    #        mask[b, i, np.random.choice(self.config.n_features, (int(self.feature_probability.cpu().numpy()[i].item()),), replace=False)] = 1
    #mask = torch.Tensor(mask).to(self.encoder.device)
    batch = torch.where(
        torch.rand((n_batch, self.config.n_instances, self.config.n_features), device=self.encoder.device) <= self.feature_probability,
        feat,
        torch.zeros((), device=self.encoder.device),
    )
    return batch
```

```python
def linear_lr(step, steps):
  return (1 - (step / steps))

def constant_lr(*_):
  return 1.0

def cosine_decay_lr(step, steps):
  return np.cos(0.5 * np.pi * step / (steps - 1))

def optimize(model, 
             render=False, 
             n_batch=1024,
             steps=10_000,
             print_freq=100,
             lr=1e-3,
             lr_scale=constant_lr,
             hooks=[]):
  cfg = model.config

  opt = torch.optim.AdamW(list(model.parameters()), lr=lr)

  start = time.time()
  with trange(steps) as t:
    for step in t:
      step_lr = lr * lr_scale(step, steps)
      for group in opt.param_groups:
        group['lr'] = step_lr
      opt.zero_grad(set_to_none=True)
      batch = model.generate_batch(n_batch)
      out = model(batch)
      ground_truth = model.ground_truth(batch)
      error = (model.importance*(ground_truth - out)**2)
      loss = einops.reduce(error, 'b i f -> i', 'mean').sum()
      loss.backward()
      opt.step()
    
      if hooks:
        hook_data = dict(model=model,
                         step=step, 
                         opt=opt,
                         error=error,
                         loss=loss,
                         lr=step_lr)
        for h in hooks:
          h(hook_data)
      if step % print_freq == 0 or (step + 1 == steps):
        t.set_postfix(
            loss=loss.item() / cfg.n_instances,
            lr=step_lr,
        )
        
        
def evaluate(model, 
             render=False, 
             n_batch=10000,
             steps=100,
             print_freq=100,
             lr=1e-3,
             lr_scale=constant_lr,
             hooks=[]):
  cfg = model.config

  opt = torch.optim.AdamW(list(model.parameters()), lr=lr)

  start = time.time()


  batch = model.generate_batch(n_batch)
  out = model(batch)
  ground_truth = model.ground_truth(batch)
  error = (model.importance*(ground_truth - out)**2)
  loss = einops.reduce(error, 'b i f -> i', 'mean')
    
  baseline_error = (model.importance*(out.mean((0,), keepdims=True) - out)**2)
  variance = einops.reduce(baseline_error, 'b i f -> i', 'mean')

  return loss / variance
```

```python
if torch.cuda.is_available():
  DEVICE = 'cuda'
else:
  DEVICE = 'cpu'
```

```python
models = {}
losses = {}
```

```python
n_features_options = [64, 128, 256, 512, 1024]
n_features_to_fit_options = [1]+n_features_options
n_neurons_options = [128]
n_layers_options = [0, 1, 2]
n_layers_ground_truth_options = [0, 1]
n_sparsity = 5
n_trials = 10
```

```python

for n_layers_ground_truth in n_layers_ground_truth_options:
    for n_features in n_features_options:
        for n_features_to_fit in n_features_to_fit_options:
            if n_features_to_fit not in [1, n_features]:
                continue
            for n_neurons in n_neurons_options:
                for n_layers in n_layers_options:
                    #if n_layers == n_layers_ground_truth:
                    #    continue
                    print("In iteration", (n_layers_ground_truth, n_features, n_features_to_fit, n_neurons, n_layers))


                    config = Config(
                        n_features = n_features,
                        n_neurons = n_neurons,
                        n_layers = n_layers,
                        n_layers_ground_truth = n_layers_ground_truth,
                        n_features_to_fit = n_features_to_fit,
                        n_instances = n_sparsity*n_trials,
                    )

                    model = Model(
                        config=config,
                        device=DEVICE,
                        # Exponential feature importance curve from 1 to 1/100
                        importance = torch.ones(config.n_features_to_fit)[None, :],
                        feature_probability = torch.cat([(2 ** -torch.linspace(0, 8, n_sparsity)) for jj in range(n_trials)], dim=0)[:, None]
                        #feature_probability = torch.cat([(2 ** torch.linspace(0, 3, n_sparsity)) for jj in range(n_trials)], dim=0)[:, None]

                    )

                    optimize(model, steps=50000, lr=1e-3, print_freq=100)

                    models[(n_layers_ground_truth, n_features, n_features_to_fit, n_neurons, n_layers)] = model

                    loss = evaluate(model)
                    losses[(n_layers_ground_truth, n_features, n_features_to_fit, n_neurons, n_layers)] = loss.detach().cpu().numpy()

                    np.save(save_dir+"/Zlosses.npy", losses)

                    np.save(save_dir+"/Zloss_nfeat_"+str(n_features)+("_nfeat2fit_")+str(n_features_to_fit)+"_nneurons_"+str(n_neurons)+"_n_layers_"+str(n_layers)+"_n_layers_gt_"+str(n_layers_ground_truth)+".npy",
                           loss.detach().cpu().numpy())

                    torch.save(model, save_dir+"/Zmodel_nfeat_"+str(n_features)+("_nfeat2fit_")+str(n_features_to_fit)+"_nneurons_"+str(n_neurons)+"_n_layers_"+str(n_layers)+"_n_layers_gt_"+str(n_layers_ground_truth)+".pth",)
```
